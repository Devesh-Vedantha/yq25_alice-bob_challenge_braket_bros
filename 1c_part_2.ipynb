### 2-Cat State

We see the characteristic fringes (whiskers) of the cat ! The two blobs are defined up to a phase so that is ok that it's rotated. Quantum mechanics doesn't care about global phases

```python
rho_file_path = "2_cat_rho.npy" # path to pickle file containing new rho, output from 1b;
rho = np.load(rho_file_path);
```

STATE IS VERY PURE!

```python
print(np.real(np.trace(rho @ rho)))
```

We see due to the interference we 'cancel out' the odd fringes of the cat. We are left with a **Poisson** distribution of only the even-cat states. So we know this is an even cat with + parity protection

```python
dq.plot.fock(rho)
```
